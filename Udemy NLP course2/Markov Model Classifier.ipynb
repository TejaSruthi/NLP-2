```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
with open("edgar_allan_poe.txt") as eap:
    edgar = eap.readlines()
with open("robert_frost.txt") as rf:
    robert = rf.readlines()

# edgar = edgar.split("\n\n")
# robert = robert.split("\n\n")
```

```python
labels = []
input_data = []
for i,j in enumerate(edgar):
    j = j.rstrip().lower()
    edgar[i] = re.sub("[^a-z ]","",j)
    if len(edgar[i]) != 0:
        labels.append(0)
        input_data.append(edgar[i])
 
for i,j in enumerate(robert):
    j = j.rstrip().lower()
    robert[i] = re.sub("[^a-z ]","",j)
    if len(robert[i])!=0:
        labels.append(1)
        input_data.append(robert[i])
```

```python
len(input_data)
```

```python
X_train,X_test,y_train,y_test = train_test_split(input_data,labels,test_size=0.2,random_state=42)
```

```python
idx =1
word2idx = {"<unk>":0}
for i in X_train:
    for j in i.split():
        if j not in word2idx:
            word2idx[j]=idx
            idx+=1
```

```python
train_data_int=[]
text_data_int=[]
for i in X_train:
    train_data_int.append([word2idx.get(k,0) for k in i.split()])
for i in X_test:
    text_data_int.append([word2idx.get(k,0) for k in i.split()])
```

```python
v= len(word2idx)
```

```python
# initialize A and pi for both classes
A0 = np.ones((v,v))
pi0 = np.ones(v)

A1 = np.ones((v,v))
pi1 = np.ones(v)
```

```python
def compute_counts(data_int,A,pi):
    for tokens in data_int:
        last_idx = None
        for idx in tokens:
            if last_idx is None:
                pi[idx]+=1
            else:
                A[last_idx,idx] += 1
            last_idx = idx
            
compute_counts([k for k,y in zip(train_data_int,y_train) if y==0],A0,pi0)
compute_counts([k for k,y in zip(train_data_int,y_train) if y==1],A1,pi1)
```

```python
A0 /= A0.sum(axis=1,keepdims=True)
pi0 /= pi0.sum()

A1 /= A1.sum(axis=1,keepdims=True)
pi1 /= pi1.sum()
```

```python
logA0 = np.log(A0)
logpi0 = np.log(pi0)

logA1 = np.log(A1)
logpi1 = np.log(pi1)
```

```python
count0 = sum(y == 0 for y in y_train)
count1 = sum(y == 1 for y in y_train)
total = len(y_train)
p0 = count0/total
p1 = count1/total
logp0 = np.log(p0)
logp1 = np.log(p1)
p0,p1
```

```python
class Classifier:
    def __init__(self, logAs, logpis, logpriors):
        self.logAs = logAs
        self.logpis = logpis
        self.logpriors = logpriors
        self.K = len(logpriors)

    def _compute_log_likelihood(self, input_, class_):
        logA = self.logAs[class_]
        logpi = self.logpis[class_]
        last_idx = None
        logprob =0
        for idx in input_:
            if last_idx is None:
                logprob +=logpi[idx]
            else:
                logprob += logA[last_idx,idx]
            last_idx = idx

        return logprob

    def predict(self, inputs):
        predictions = np.zeros(len(inputs))
        for i, input_ in enumerate(inputs):
            posteriors = [self._compute_log_likelihood(input_,c) + self.logpriors[c] for c in range(self.K)]
            pred = np.argmax(posteriors)
            predictions[i] = pred
        return predictions
```

```python
clf = Classifier([logA0, logA1], [logpi0, logpi1], [logp0, logp1])
```

```python
Ptrain = clf.predict(train_data_int)
print(f"Test accuracy: {np.mean(Ptrain==y_train)*100:.2f}%")
```

```python
from sklearn.metrics import confusion_matrix, f1_score
cm = confusion_matrix(y_train, Ptrain)
cm
```
