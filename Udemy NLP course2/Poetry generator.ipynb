```python
import numpy as np
import re
from collections import Counter
np.random.seed(1234)
```

```python
initial={}
first_order ={}
second_order ={}
```

```python
def add2dict(d,k,v):
    if k not in d:
        d[k] = []
    d[k].append(v)
```

```python
for line in open("robert_frost.txt"):
    line = line.lower()
    line = re.sub("[^a-z ]","",line)
    if len(line)!=0:
        line = line.rstrip().split()
        n = len(line)
        for i in range(n):
            t = line[i]
            if i==0:
                initial[t] = initial.get(t,0.)+1
            else:
                t_1 = line[i-1]
                if i == n-1:
                    add2dict(second_order, (t_1,t), "END")
                if i == 1:
                    add2dict(first_order, t_1, t)
                else:
                    t_2 = line[i-2]
                    add2dict(second_order, (t_2,t_1), t)
    
```

```python
initial_total = sum(initial.values())
for t,c in initial.items():
    initial[t] = c/initial_total
```

```python
def list2pdict(ts):
    d={}
    n = len(ts)
    for t in ts:
       d[t] = d.get(t,0.)+1
    for t,c in d.items():
        d[t] = c/n
    return d
```

```python
for t_1, ts in first_order.items():
    first_order[t_1] = list2pdict(ts)
```

```python
for k,ts in second_order.items():
    second_order[k] = list2pdict(ts)
```

```python
def sample_word(d):
    p0 = np.random.random()
    cumulative =0
    for t,p in d.items():
        cumulative += p
        if p0<cumulative:
            return t
    assert(False)
```

```python
def generate():
    for i in range(4):
        sentence =[]
        w0 = sample_word(initial)
        sentence.append(w0)
        w1 = sample_word(first_order[w0])
        sentence.append(w1)

        while (True):
            w2 = sample_word(second_order[(w0,w1)])
            if w2=='END':
                break
            sentence.append(w2)
            w0 = w1
            w1 = w2
        print(" ".join(sentence))
```
